3. Given the 2D array:
scores = np.array([[85, 78, 92],
                   [90, 88, 76],
                   [72, 95, 89],
                   [88, 82, 91]])
                   
        a. Reverse the entire array using two different methods.
        b. Extract the second row (index 1).
        c. Reverse all rows (flip upside down).

```python
import numpy as np
scores = ([
        [90,88,76],
        [72,95,89],
        [88,82,91]
 ])
print("original array")
print(scores)
method_a=scores[::-1]
print("reverse the entire array method(a):")
print(method_a)
method_b=np.flip(scores)
print("reverse the entire array method(b):")
print(method_b)
second_row=[1]
print("second row:")
print(second_row)
reverse_all_rows=np.flipud(scores)
print("reverse all rows")
print(reverse_all_rows)
```
