Array properties:

```python
import numpy as np
arr = np.array([[1,2,3],[4,5,6]])
print("original array",arr)
print("shape",arr.shape)
print("size",arr.size)
print("dimentional",arr.ndim)
print("data type",arr.dtype)
print("transpose",arr.T)
print("type",type(arr))
print("own data",arr.flags['OWNDATA'])
print("list",arr.tolist())
```
